```python
import pandas as pd
import numpy as np
import time

# SecML imports
from secml.array import CArray
from secml.data import CDataset
from secml.data.splitter import CTrainTestSplit
from secml.data.splitter import CDataSplitterShuffle

# Qiskit imports cho VQC
from qiskit.circuit.library import ZZFeatureMap, RealAmplitudes # Th√™m RealAmplitudes (Ansatz)
from qiskit import BasicAer
from qiskit.utils import QuantumInstance
from qiskit_machine_learning.algorithms import VQC # ƒê·ªïi t·ª´ QSVC sang VQC
from qiskit.algorithms.optimizers import COBYLA # Th√™m Optimizer (B·ªô t·ªëi ∆∞u)
from sklearn.preprocessing import OneHotEncoder # C·∫ßn thi·∫øt cho VQC

print("‚úÖ ƒê√£ import xong th∆∞ vi·ªán cho VQC!")
```

```python
# ==========================================
# LOAD D·ªÆ LI·ªÜU + CH·ªåN FEATURES (ƒê√É T·ªêI ∆ØU T·ªêC ƒê·ªò)
# ==========================================
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, chi2
from secml.data.splitter import CTrainTestSplit
from secml.data import CDataset
from secml.array import CArray

# 1. ƒê·ªçc file CSV
csv_file = 'drebin-215-dataset-5560malware-9476-benign.csv'
print(f"Loading dataset: {csv_file}...")
df = pd.read_csv(csv_file)

# Fix l·ªói k√Ω t·ª± l·∫°
df = df.replace('?', np.nan).fillna(0)

# X·ª≠ l√Ω nh√£n
if df.iloc[:, -1].dtype == object:
    df.iloc[:, -1] = df.iloc[:, -1].map({'S': 1, 'B': 0}).fillna(0).astype(int)

X_full = df.iloc[:, :-1].values.astype(np.float64)
y_full = df.iloc[:, -1].values.astype(int)

# ---------------------------------------------------------
# C·∫¨P NH·∫¨T: GI·∫¢M XU·ªêNG 8 QUBITS ƒê·ªÇ TRAIN NHANH
# ---------------------------------------------------------
LIMIT_QUBITS = 12  # <--- ƒê√É S·ª¨A T·ª™ 20 XU·ªêNG 8
print(f"‚ö†Ô∏è ƒêang ch·ªçn l·ªçc {LIMIT_QUBITS} features t·ªët nh·∫•t (T·ªëi ∆∞u t·ªëc ƒë·ªô)...")

selector = SelectKBest(chi2, k=LIMIT_QUBITS)
X_selected = selector.fit_transform(X_full, y_full)

# T·∫°o Dataset v√† chia m·∫´u
ds = CDataset(CArray(X_selected), CArray(y_full))
labels = ds.Y.tondarray().ravel()

# L·∫•y 250 m·∫´u l√†nh t√≠nh v√† 250 m·∫´u ƒë·ªôc h·∫°i -> T·ªïng 500 m·∫´u
idx_benign = np.where(labels == 0)[0][:250]
idx_malware = np.where(labels == 1)[0][:250]

selected_indices = np.concatenate([idx_benign, idx_malware])
ds_500 = ds[selected_indices.tolist(), :]

# X√°o tr·ªôn d·ªØ li·ªáu ng·∫´u nhi√™n
shuffled_indices = np.random.permutation(ds_500.num_samples)
ds_500 = ds_500[shuffled_indices.tolist(), :]

# ---------------------------------------------------------
# SPLIT 400 TRAIN - 100 TEST
# ---------------------------------------------------------
tr, ts = CTrainTestSplit(train_size=400, random_state=42).split(ds_500)

print("-" * 30)
print(f"T·ªïng s·ªë m·∫´u: {ds_500.num_samples}")
print(f"S·ªë l∆∞·ª£ng Train: {tr.num_samples}")
print(f"S·ªë l∆∞·ª£ng Test: {ts.num_samples}")
print(f"S·ªë l∆∞·ª£ng Features (Qubits): {ds_500.num_features}") # S·∫Ω hi·ªán l√† 8
print("-" * 30)
```

```python
# ==========================================
# C·∫§U H√åNH VQC (S·ª¨ D·ª§NG SAMPLER ƒê·ªÇ L·∫§Y X√ÅC SU·∫§T)
# ==========================================
from qiskit.primitives import Sampler # Import m·ªõi
from qiskit_machine_learning.algorithms import VQC
from qiskit.algorithms.optimizers import COBYLA
from qiskit.circuit.library import ZZFeatureMap, RealAmplitudes

# 1. Feature Map (M√£ h√≥a d·ªØ li·ªáu)
num_features = ds_500.num_features 
feature_map = ZZFeatureMap(feature_dimension=num_features, reps=2)

# 2. Ansatz (M·∫°ch bi·∫øn ph√¢n)
ansatz = RealAmplitudes(num_qubits=num_features, reps=2)

# 3. Optimizer
optimizer = COBYLA(maxiter=50) 

# 4. Kh·ªüi t·∫°o Model VQC v·ªõi SAMPLER (Thay cho QuantumInstance)
# Sampler gi√∫p t√≠nh to√°n x√°c su·∫•t (quasi-probabilities) thay v√¨ ch·ªâ ƒëo ƒë·∫°c shot
sampler = Sampler() 

vqc = VQC(feature_map=feature_map,
          ansatz=ansatz,
          optimizer=optimizer,
          sampler=sampler) # D√πng tham s·ªë sampler thay v√¨ quantum_instance

print("‚úÖ ƒê√£ kh·ªüi t·∫°o VQC v·ªõi Sampler! (S·∫µn s√†ng xu·∫•t x√°c su·∫•t)")
```

```python
# ==========================================
# HU·∫§N LUY·ªÜN VQC (L∆∞u √Ω: C·∫ßn One-Hot Label)
# ==========================================
start_time = time.time()
print("Training VQC started (Qu√° tr√¨nh n√†y c√≥ th·ªÉ l√¢u h∆°n QSVC)...")

# 1. Chu·∫©n b·ªã d·ªØ li·ªáu Numpy
X_train_np = tr.X.tondarray()
y_train_np = tr.Y.tondarray().ravel()

X_test_np = ts.X.tondarray()
y_test_np = ts.Y.tondarray().ravel()

# 2. One-Hot Encoding cho nh√£n (B·∫Øt bu·ªôc ƒë·ªëi v·ªõi VQC trong Qiskit ƒë·ªÉ t·ªëi ∆∞u t·ªët nh·∫•t)
# Nh√£n 0 -> [1, 0], Nh√£n 1 -> [0, 1]
encoder = OneHotEncoder(sparse=False)
y_train_oh = encoder.fit_transform(y_train_np.reshape(-1, 1))
y_test_oh = encoder.transform(y_test_np.reshape(-1, 1))

# 3. Hu·∫•n luy·ªán (Fit)
# VQC fit v·ªõi nh√£n one-hot
vqc.fit(X_train_np, y_train_oh)
print("Training completed!")

# 4. ƒê√°nh gi√° s∆° b·ªô
# score trong VQC tr·∫£ v·ªÅ mean accuracy
train_score = vqc.score(X_train_np, y_train_oh)
test_score = vqc.score(X_test_np, y_test_oh)

print("-" * 30)
print(f"Accuracy on Train set: {train_score:.2f}")
print(f"Accuracy on Test set:  {test_score:.2f}")
print(f"Total execution time: {time.time() - start_time:.2f} seconds")
print("-" * 30)
```

```python
# ==========================================
# ƒê√ÅNH GI√Å CHI TI·∫æT (UPDATE THEO Y√äU C·∫¶U)
# ==========================================
from sklearn.metrics import roc_curve, auc, f1_score, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import numpy as np

print("üîç ƒêang ti·∫øn h√†nh d·ª± ƒëo√°n v√† t√≠nh to√°n x√°c su·∫•t...")

# ---------------------------------------------------------
# 1. D·ª∞ ƒêO√ÅN (PREDICT) & L·∫§Y X√ÅC SU·∫§T (PROBABILITY)
# ---------------------------------------------------------
# L·∫•y nh√£n d·ª± ƒëo√°n (0 ho·∫∑c 1) ƒë·ªÉ t√≠nh Accuracy/F1
# L∆∞u √Ω: N·∫øu vqc tr·∫£ v·ªÅ one-hot, ta c·∫ßn convert l·∫°i (t√πy phi√™n b·∫£n qiskit)
try:
    # C√°ch chu·∫©n cho Qiskit Machine Learning m·ªõi
    y_pred = vqc.predict(X_test_np)
    # L·∫•y x√°c su·∫•t (Confidence Score) ƒë·ªÉ v·∫Ω ROC
    # predict_proba tr·∫£ v·ªÅ [[xs_l·ªõp_0, xs_l·ªõp_1], ...] -> ta l·∫•y c·ªôt 1
    y_probs = vqc.predict_proba(X_test_np)
    y_score = y_probs[:, 1] 
except:
    # Fallback cho m·ªôt s·ªë tr∆∞·ªùng h·ª£p c≈© ho·∫∑c c·∫•u h√¨nh ƒë·∫∑c bi·ªát
    print("‚ö†Ô∏è C·∫£nh b√°o: ƒêang s·ª≠ d·ª•ng fallback mode cho predict...")
    y_pred_oh = vqc.predict(X_test_np)
    y_pred = encoder.inverse_transform(y_pred_oh).ravel()
    # N·∫øu kh√¥ng c√≥ predict_proba, t·∫°m d√πng ch√≠nh one-hot output l√†m score
    # (Ch·ªâ ƒë√∫ng n·∫øu vqc.predict tr·∫£ v·ªÅ soft-probability, n·∫øu tr·∫£ v·ªÅ c·ª©ng 0/1 th√¨ bi·ªÉu ƒë·ªì s·∫Ω b·ªã vu√¥ng)
    y_score = y_pred_oh[:, 1] if len(y_pred_oh.shape) > 1 else y_pred

# ---------------------------------------------------------
# 2. T√çNH TO√ÅN C√ÅC CH·ªà S·ªê (Metrics)
# ---------------------------------------------------------
# C√°c ch·ªâ s·ªë c∆° b·∫£n
acc = accuracy_score(y_test_np, y_pred)
f1 = f1_score(y_test_np, y_pred)

# T√≠nh ƒë∆∞·ªùng cong ROC
fpr, tpr, thresholds = roc_curve(y_test_np, y_score)
roc_auc = auc(fpr, tpr)

# --- Logic t√¨m Detection Rate t·∫°i 2% FPR (Gi·ªëng Cell 13) ---
target_fpr = 0.02  # M·ªëc 2%
# T√¨m v·ªã tr√≠ index m√† t·∫°i ƒë√≥ FPR g·∫ßn 0.02 nh·∫•t
idx = (np.abs(fpr - target_fpr)).argmin()
dr_at_2fpr = tpr[idx] # Detection Rate t·∫°i m·ªëc ƒë√≥

# ---------------------------------------------------------
# 3. IN K·∫æT QU·∫¢
# ---------------------------------------------------------
print("-" * 40)
print(f"üìä K·∫æT QU·∫¢ ƒê√ÅNH GI√Å TR√äN {len(y_test_np)} M·∫™U TEST:")
print(f"‚úÖ Accuracy:               {acc:.2%}")
print(f"‚úÖ F1 Score:               {f1:.2%}")
print("-" * 40)
print(f"üéØ K·∫æT QU·∫¢ T·∫†I NG∆Ø·ª†NG AN TO√ÄN (FPR ~ 2%):")
print(f"   - False Positive Rate:  {fpr[idx]:.2%} (M·ª•c ti√™u: 2%)")
print(f"   - Detection Rate (DR):  {dr_at_2fpr:.2%}")
print("-" * 40)
print("Confusion Matrix:")
print(confusion_matrix(y_test_np, y_pred))

# ---------------------------------------------------------
# 4. V·∫º BI·ªÇU ƒê·ªí ROC (Gi·ªëng Cell 14)
# ---------------------------------------------------------
plt.figure(figsize=(8, 5))

# V·∫Ω ƒë∆∞·ªùng ROC
plt.plot(fpr, tpr, color='darkgreen', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')

# V·∫Ω ƒë∆∞·ªùng tham chi·∫øu n√©t ƒë·ª©t t·∫°i FPR = 2% (Tr·ª•c d·ªçc)
plt.axvline(x=target_fpr, color='red', linestyle='--', label='2% FPR Threshold')

# Trang tr√≠ bi·ªÉu ƒë·ªì
plt.xlim([0.0, 0.1])  # Zoom v√†o v√πng 0% - 10% FPR (Quan tr·ªçng!)
plt.ylim([0.0, 1.05]) # Tr·ª•c Y t·ª´ 0 ƒë·∫øn 100%
plt.xlabel('False Positive Rate (T·ª∑ l·ªá b√°o ƒë·ªông gi·∫£)')
plt.ylabel('Detection Rate (T·ª∑ l·ªá ph√°t hi·ªán)')
plt.title('ROC Curve @ Low False Positive Rate')
plt.legend(loc="lower right")
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
# ==========================================
# KI·ªÇM TRA M·∫™U TH·ª∞C T·∫æ T·ª™ DATASET (Blind Test)
# ==========================================
import numpy as np

# 1. C·∫•u h√¨nh
SAMPLES_TO_CHECK = 10  # S·ªë l∆∞·ª£ng m·∫´u mu·ªën ki·ªÉm tra ng·∫´u nhi√™n
current_n_features = X_test_np.shape[1]

print(f"‚ÑπÔ∏è Model VQC (Input {current_n_features} features) - Ki·ªÉm tra {SAMPLES_TO_CHECK} m·∫´u ng·∫´u nhi√™n:\n")

# 2. Ch·ªçn ng·∫´u nhi√™n index t·ª´ t·∫≠p Test
# replace=False ƒë·ªÉ kh√¥ng ch·ªçn tr√πng
random_indices = np.random.choice(len(X_test_np), SAMPLES_TO_CHECK, replace=False)

# 3. V√≤ng l·∫∑p ki·ªÉm tra
correct_count = 0

print(f"{'STT':<5} | {'D·ª± ƒëo√°n (AI)':<20} | {'Th·ª±c t·∫ø (Label)':<20} | {'K·∫øt qu·∫£':<10}")
print("-" * 70)

for i, idx in enumerate(random_indices):
    # L·∫•y features v√† nh√£n th·ª±c t·∫ø
    sample_features = X_test_np[idx]
    true_label_code = y_test_np[idx]
    
    # --- D·ª∞ ƒêO√ÅN ---
    input_np = sample_features.reshape(1, -1)
    
    # Predict (tr·∫£ v·ªÅ one-hot)
    pred_oh = vqc.predict(input_np)
    # Decode one-hot v·ªÅ nh√£n 0/1
    pred_label_code = encoder.inverse_transform(pred_oh).ravel()[0]
    
    # --- HI·ªÇN TH·ªä ---
    # Chuy·ªÉn m√£ 0/1 th√†nh ch·ªØ
    pred_text = "üíÄ M√É ƒê·ªòC" if pred_label_code == 1 else "‚úÖ L√ÄNH T√çNH"
    true_text = "üíÄ M√É ƒê·ªòC" if true_label_code == 1 else "‚úÖ L√ÄNH T√çNH"
    
    # So s√°nh
    if pred_label_code == true_label_code:
        result_icon = "‚úîÔ∏è ƒê√öNG"
        correct_count += 1
    else:
        result_icon = "‚ùå SAI"

    print(f"#{idx:<4} | {pred_text:<20} | {true_text:<20} | {result_icon:<10}")

print("-" * 70)
print(f"üéØ ƒê·ªô ch√≠nh x√°c tr√™n {SAMPLES_TO_CHECK} m·∫´u th·ª≠: {correct_count}/{SAMPLES_TO_CHECK} ({correct_count/SAMPLES_TO_CHECK:.0%})")
```
